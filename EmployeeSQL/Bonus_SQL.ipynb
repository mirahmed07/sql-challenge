```python
# Import the Employees table
employees = pd.read_sql('select * from employees', connection, parse_dates=['birth_date', 'hire_date'])
employees.head()
```

```python
# Import the Departments table
departments = pd.read_sql('select * from departments', connection)
departments.head()
```

```python
# Import the Salaries table
salaries = pd.read_sql('select * from salaries', connection, parse_dates=['from_date', 'to_date'])
salaries.head()
```

```python
# Import the Department Manager table
dept_manager = pd.read_sql('select * from dept_manager', connection, parse_dates=['from_date', 'to_date'])
dept_manager.head()
```

```python
# Create a merged dataframe of titles and salaries
employee_salaries = pd.merge(employees, salaries, on = 'emp_no')
employee_salaries = employee_salaries.merge(titles, left_on = 'emp_title', right_on = 'title_id')
employee_salaries.sort_values(by = 'emp_no', inplace =True)
employee_salaries.reset_index(drop= True, inplace =True)
del employee_salaries['title_id']
employee_salaries.head()
```

```python
# Creating Dataframe with emp_no, title, salary
employee_salaries_df = employee_salaries[['emp_no', 'title', 'salary']].reset_index(drop = True)
employee_salaries_df.sort_values(by = 'emp_no', inplace =True)
employee_salaries_df.reset_index(drop= True, inplace =True)
employee_salaries_df.head()
```

## creating bins for salary ranges

```python
bins = [employee_salaries_df["salary"].min() ,50000, 60000, 70000, 80000, 90000, 100000, 110000, employee_salaries_df["salary"].max()]
salary_grp = ['=<50k', '(50k+1)-60k', '(60k+1)-70k', '(70k+1)-80k', '(80k+1)-90k', '(90k+1)-100k', '(100k+1)-110k', '>110k']
employee_salaries_df["salary_grp"]= pd.cut(employee_salaries_df["salary"], bins, labels=salary_grp, include_lowest = True)
employee_salaries_df.head()
```

## Getting outliers for each title group

```python
# Put titles into a list for for loop (and later for plot labels)
title_list = list(set(titles["title"]))
# Create empty list to fill with salary (for plotting)
salary = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for title in title_list: 
    # Locate the rows for each title and get the salary
    salary_entry = employee_salaries_df["salary"].loc[employee_salaries_df["title"] == title]
    
    # add subset 
    salary.append(salary_entry)
    
    # Calculate quartiles for each title
    quartiles = salary_entry.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    
    # Print IQR for each drug
    print(f'IQR for {title}: {iqr}')
    
    # Calculate upper and lower bounds for each title
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    # Print lower and upper bounds for each title
    print(f'Lower Bound for {title}: {lower_bound}')
    print(f'Upper Bound for {title}: {upper_bound}')
    
    # Determine outliers using upper and lower bounds
    outliers_count = (employee_salaries_df.loc[(employee_salaries_df['salary'] >= upper_bound) | 
                                        (employee_salaries_df['salary'] <= lower_bound)]).count()
    
    # Print outliers for each drug
    print(f'Number of {title} outliers: {outliers_count}')
```

# linear regression for experience on years and salaries

```python
# Define function for calculating linear regression and scatter plot with linear regression plot
def linear_regression(x,y,title):

    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    print(f"The correlation coefficient of {title} is : {rvalue}")
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x, y, c = "lightblue", edgecolor = "black")
    plt.plot(x,regress_values,"r-")
    plt.title(f"{title}'s correlation value: {rvalue}")
    plt.ylabel("Salary")
    plt.xlabel("Experience in years")
    plt.grid(alpha = .2)
    return line_eq

# Define a fuction for annotating
def plot_annotate(line_eq, a, b):
    plt.annotate(line_eq,(a,b),fontsize=15,color="black")
```

```python
def ln_rg_title(title):
    
    x = employee_salaries["experience"].loc[employee_salaries["title"] == title]
    y = employee_salaries["salary"].loc[employee_salaries["title"] == title]
    a = x.min()
    b = y.max()
    line_eq = linear_regression(x,y, title)
    
    plot_annotate(line_eq, a, b)
   
    plt.show()
```
